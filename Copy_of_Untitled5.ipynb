```python
#import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 5)
```

```python
# 1. Line Chart - Total number sold over time
df.groupby('Date')['number_sold'].sum().plot()
plt.title('Total Number Sold Over Time')
plt.xlabel('Date')
plt.ylabel('Number Sold')
plt.grid(True)
plt.show()
```

```python
# 2. Scatter Plot - Store vs Number Sold
sns.scatterplot(x='store', y='number_sold', data=df, alpha=0.3)
plt.title('Store vs Number Sold')
plt.show()
```

```python
# 3. Box Plot - Distribution of number_sold per store
sns.boxplot(x='store', y='number_sold', data=df)
plt.title('Box Plot of Number Sold per Store')
plt.show()
```

```python
# 4. Heatmap - Correlation matrix
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# 5. Barplot - Average sales per product
avg_sales = df.groupby('product')['number_sold'].mean().reset_index()
sns.barplot(x='product', y='number_sold', data=avg_sales, palette="viridis")
plt.title('Average Sales per Product')
plt.xlabel('Product')
plt.ylabel('Avg Number Sold')
plt.show()
```

```python
# 6. Pie Chart - Sales share by store
sales_by_store = df.groupby('store')['number_sold'].sum()
plt.pie(sales_by_store, labels=sales_by_store.index, autopct='%1.1f%%', startangle=140)
plt.title('Sales Distribution by Store')
plt.axis('equal')
plt.show()
```

```python
# 7. Countplot - Number of records per store
sns.countplot(x='store', data=df)
plt.title('Number of Records per Store')
plt.show()
```

```python
# 8. Violinplot - Distribution of number_sold per product
sns.violinplot(x='product', y='number_sold', data=df, inner='quartile')
plt.title('Violin Plot of Number Sold per Product')
plt.show()
```

```python
# 9. Histogram - Distribution of number sold
sns.histplot(df['number_sold'], bins=30, kde=True)
plt.title('Distribution of Number Sold')
plt.xlabel('Number Sold')
plt.show()
```

```python
# 10. Pair Plot - Numeric relationships
sns.pairplot(df[['store', 'product', 'number_sold']])
plt.suptitle('Pair Plot of Numeric Columns', y=1.02)
plt.show()
```
